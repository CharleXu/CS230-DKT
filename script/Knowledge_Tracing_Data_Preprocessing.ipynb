```python
def get_groups(skill_nm):
    for k,v in groups.items():
        if skill_nm in v:
            return k
        
def get_attempt(x):
    if x>5:
        return '>5'
    elif x>2:
        return '3-5'
    else:
        return str(int(x))
```

```python
data['skill_group'] = data['skill_name'].apply(lambda x:get_groups(x))
data['attempt'] = data['attempt_count'].apply(lambda x:get_attempt(x))
```
